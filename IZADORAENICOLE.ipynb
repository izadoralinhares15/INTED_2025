```python
#2 – Uma escola deseja automatizar a verificação do desempenho dos alunos. Leia duas notas, calcule a média e utilize uma estrutura if…else para informar se o aluno está aprovado (média ≥ 7) ou reprovado.

nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))

media = (nota1 + nota2) / 2

if media >= 7:
    print(f"Média: {media:.2f}. Aluno APROVADO.")
else:
    print(f"Média: {media:.2f}. Aluno REPROVADO.")
```

```python
#3 – Um sistema meteorológico quer classificar a sensação térmica do dia. A partir de uma temperatura informada, determine se está fria, agradável, quente ou muito quente usando if…elif…else.

temperatura = float(input("Informe a temperatura em °C: "))

if temperatura < 15:
    print("Sensação térmica: FRIA.")
elif 15 <= temperatura < 25:
    print("Sensação térmica: AGRADÁVEL.")
elif 25 <= temperatura < 35:
    print("Sensação térmica: QUENTE.")
else:
    print("Sensação térmica: MUITO QUENTE.")
```

```python
#5 – Uma empresa deseja reajustar salários com base em faixas de rendimento. Receba o salário atual e aplique um aumento conforme as faixas (10%, 7% ou 5%), usando condicionais encadeadas. Exiba o novo salário.

salario_atual = float(input("Informe o salário atual: R$ "))

if salario_atual <= 1500:
    percentual_aumento = 0.10
elif salario_atual <= 3000:
    percentual_aumento = 0.07
else:
    percentual_aumento = 0.05

aumento = salario_atual * percentual_aumento
novo_salario = salario_atual + aumento

print(f"Aumento aplicado: {percentual_aumento * 100:.0f}%")
print(f"Novo salário: R$ {novo_salario:.2f}")
```

```python
#11 – Uma calculadora avançada deve permitir que o usuário some quantos números quiser. Peça números repetidamente até que o valor 0 seja digitado. Ao final, mostre a soma total.

soma_total = 0
print("Digite números para somar. Digite 0 para encerrar.")

while True:
    try:
        numero = float(input("Número: "))
        if numero == 0:
            break  5254
        soma_total += numero
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")

print(f"A soma total dos números digitados é: {soma_total}")
```

```python

#13 – Uma máquina de simulação deve somar números sequenciais começando em 1 até que a soma ultrapasse 100, mostrando cada número utilizado. Utilize um laço while.

soma = 0
numero_atual = 1

while soma <= 100:
    soma += numero_atual
    print(f"Número utilizado: {numero_atual}, Soma atual: {soma}")
    numero_atual += 1
print(f"A soma final ultrapassou 100, terminando em: {soma}")
```

```python

#14 – Uma pesquisa está registrando idades. Continue lendo idades até que o usuário digite um valor negativo. Ao final, calcule e exiba a média das idades digitadas.

soma_idades = 0
contador_idades = 0

print("Digite as idades. Digite um valor negativo para encerrar.")

while True:
    try:
        idade = int(input("Idade: "))
        if idade < 0:
            break

        soma_idades += idade
        contador_idades += 1
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

print("-" * 20)

if contador_idades > 0:
    media = soma_idades / contador_idades
    print(f"Total de idades digitadas: {contador_idades}")
    print(f"Média das idades: {media:.2f}")
else:
    print("Nenhuma idade válida foi digitada.")
```

```python
#15 – Um cronômetro digital deve realizar uma contagem regressiva personalizada. Receba um número inicial e exiba a contagem até 0 usando um while.

try:
    contador = int(input("Digite o número inicial da contagem regressiva: "))

    if contador < 0:
        print("O número inicial deve ser não negativo.")
    else:
        while contador >= 0:
            print(contador)
            contador -= 1
        print("Contagem regressiva finalizada!")
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

```python
#19 – Em um formulário digital, serão digitadas 5 idades. Use um laço for para contar quantas delas são maiores que 18.

contador_maiores = 0
NUMERO_IDADES = 5

print(f"Digite {NUMERO_IDADES} idades:")

for i in range(NUMERO_IDADES):
    try:
        idade = int(input(f"Idade #{i + 1}: "))
        if idade > 18:
            contador_maiores += 1
    except ValueError:
        print("Idade inválida. Esta entrada será ignorada.")

print(f"\nTotal de idades maiores que 18: {contador_maiores}")
```

```python
#20 – Um sistema estatístico deve analisar números fornecidos pelo usuário. Leia 7 números e determine qual deles é o maior.

NUMERO_DE_VALORES = 7
maior_numero = None

print(f"Digite {NUMERO_DE_VALORES} números:")

for i in range(NUMERO_DE_VALORES):
    try:
        numero = float(input(f"Número #{i + 1}: "))

        if i == 0 or numero > maior_numero:
            maior_numero = numero

    except ValueError:
        print("Entrada inválida. Tente novamente com um número válido.")

if maior_numero is not None:
    print(f"\nO maior número digitado foi: {maior_numero}")
else:
    print("Não foi possível determinar o maior número (nenhuma entrada válida).")
```

```python
r#21 – Um app de produtividade quer identificar qual de quatro tarefas possui o nome mais longo. Leia quatro nomes de tarefas e exiba aquela com mais caracteres.

NUMERO_TAREFAS = 4
tarefa_mais_longa = ""
max_comprimento = -1

print(f"Digite o nome de {NUMERO_TAREFAS} tarefas:")

for i in range(NUMERO_TAREFAS):
    nome_tarefa = input(f"Tarefa #{i + 1}: ")

    comprimento_atual = len(nome_tarefa)

    if comprimento_atual > max_comprimento:
        max_comprimento = comprimento_atual
        tarefa_mais_longa = nome_tarefa

print(f"\nA tarefa com o nome mais longo é: **{tarefa_mais_longa}** (com {max_comprimento} caracteres).")
```

```python
#23 – Um módulo de análise precisa classificar uma lista de valores. Leia 10 números e exiba quantos são positivos, negativos e quantos são iguais a zero.

contador_pos = 0
contador_neg = 0
contador_zero = 0
NUMERO_VALORES = 10

print(f"Digite {NUMERO_VALORES} números:")

for i in range(NUMERO_VALORES):
    try:
        numero = float(input(f"Valor #{i + 1}: "))

        if numero > 0:
            contador_pos += 1
        elif numero < 0:
            contador_neg += 1
        else:
            contador_zero += 1
    except ValueError:
        print("Entrada inválida. Esta entrada será ignorada.")

print("-" * 20)
print(f"Resumo da Classificação:")
print(f"Positivos: {contador_pos}")
print(f"Negativos: {contador_neg}")
print(f"Iguais a zero: {contador_zero}")
```

```python
#25 – Uma caixa registradora digital permite que o usuário digite valores até inserir um código especial. Continue lendo números até que o valor 999 seja digitado. Use break para encerrar.

soma_valores = 0

print("Digite valores para a caixa registradora. Digite 999 para encerrar.")

while True:
    try:
        valor = float(input("Valor do produto: R$ "))

        if valor == 999:
            break

        if valor < 0:
            print("Valor inválido (negativo) e ignorado.")
            continue

        soma_valores += valor
    except ValueError:
        print("Entrada inválida. Digite um número.")

print(f"\nCaixa fechada. Total a pagar: R$ {soma_valores:.2f}")
```

```python
#27 – Crie um sistema de login com limite de tentativas. O usuário terá no máximo 3 tentativas para digitar a senha correta. Caso acerte antes, use break para encerrar o laço.

senha_correta = "python123"
MAX_TENTATIVAS = 3

for tentativa in range(1, MAX_TENTATIVAS + 1):
    senha_digitada = input(f"Tentativa {tentativa}/{MAX_TENTATIVAS} - Digite a senha: ")

    if senha_digitada == senha_correta:
        print("\nAcesso liberado! Bem-vindo(a).")
        break  # Senha correta, encerra o laço
    else:
        print("Senha incorreta.")
else:
    # Este 'else' do for é executado se o laço terminar SEM um 'break' (ou seja, as 3 tentativas falharam)
    print("\nNúmero máximo de tentativas excedido. Acesso bloqueado.")
```

```python
#28 – Desenvolva um programa que leia 10 números e some apenas os valores positivos. Caso um número negativo seja digitado, utilize continue para ignorá-lo.

soma_positivos = 0
NUMERO_LEITURAS = 10

print(f"Digite {NUMERO_LEITURAS} números para somar apenas os positivos.")

for i in range(NUMERO_LEITURAS):
    try:
        numero = float(input(f"Número #{i + 1}: "))

        if numero < 0:
            print("Número negativo ignorado.")
            continue

        soma_positivos += numero

    except ValueError:
        print("Entrada inválida. Esta entrada será ignorada.")

print(f"\nA soma de todos os números positivos digitados é: {soma_positivos:.2f}")
```

```python
#29 – Simule um acumulador que percorre números até que a soma ultrapasse 200. Quando isso ocorrer, encerre o laço usando break.

soma_acumulada = 0
LIMITE = 200
numero_atual = 1

print(f"Acumulando números até ultrapassar {LIMITE}:")

while True:

    if soma_acumulada > LIMITE:
        print(f"Soma ultrapassou {LIMITE}!")
        break

    print(f"Somando {numero_atual}. Soma atual: {soma_acumulada}")
    soma_acumulada += numero_atual
    numero_atual += 1

print(f"\nSoma final: {soma_acumulada}")
```

```python
#30 – Um programa de edição quer contar quantas palavras o usuário digitou. Continue lendo palavras até que o usuário digite “fim”. Depois, exiba quantas palavras válidas foram informadas.

contador_palavras = 0

print("Digite palavras (uma por linha). Digite **fim** para encerrar e contar.")

while True:
    palavra = input("Palavra: ").strip().lower()

    if palavra == "fim":
        break
    if palavra:
        contador_palavras += 1

print(f"\nVocê digitou **{contador_palavras}** palavras válidas.")
```
